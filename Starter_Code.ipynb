## Preprocessing

```python
# Importing our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

# Importing and reading the charity_data.csv.
import pandas as pd 
application_df_full = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df_full.head()
```

```python
# Dropping the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df = application_df_full.drop(["EIN","NAME"],1)
application_df
```

```python
# Determining the number of unique values in each column.
application_df.nunique()
```

```python
# Looking at APPLICATION_TYPE value counts for binning
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Choosing a cutoff value and create a list of application types to be replaced
application_types_to_replace = ['T9', 'T13', 'T12', 'T2', 'T25', 'T14', 'T29', 'T15', 'T17']

# Replacing in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Checking to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Looking at CLASSIFICATION value counts for binning
classification_value_counts = application_df['CLASSIFICATION'].value_counts()
classification_value_counts
```

```python
# Looking at CLASSIFICATION value counts >1
classification_value_counts[classification_value_counts > 1]
```

```python
# Choosing a cutoff value and create a list of classifications to be replaced
classifications_to_replace = classification_value_counts.index.tolist()[5:]

# Replacing in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")
    
# Checking to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()[:5]
```

```python
# Checking all categorical columns in the dataset
application_cat = application_df.dtypes[application_df.dtypes == "object"].index.tolist()
application_cat
```

```python
# Converting categorical data to numeric with `pd.get_dummies`
dummy_data = pd.get_dummies(data=application_df, columns=['APPLICATION_TYPE', 'AFFILIATION', 'CLASSIFICATION', 
                                             'USE_CASE', 'ORGANIZATION', 'SPECIAL_CONSIDERATIONS', 'INCOME_AMT', 'IS_SUCCESSFUL'])
```

```python
# Checking the number of unique values in each column
dummy_data.nunique()
```

```python
# Splitting our preprocessed data into our features and target arrays
y = dummy_data["IS_SUCCESSFUL_1"].values
X = dummy_data.drop(["IS_SUCCESSFUL_1", "IS_SUCCESSFUL_0"],1).values

# Splitting the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
# Creating a StandardScaler instances
scaler = StandardScaler()

# Fitting the StandardScaler
X_scaler = scaler.fit(X_train)

# Scaling the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Checking the dummy_data dataframe
dummy_data
```

## Compile, Train and Evaluate the Model

```python
# Defining the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
hidden_nodes_layer1 = 80
hidden_nodes_layer2 = 30

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(
    tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=43, activation="relu")
)

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="relu"))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Checking the structure of the model
nn.summary()
```

```python
# Compiling the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Training the model
fit_model = nn.fit(X_train_scaled,y_train,epochs=100)
```

```python
# Evaluating the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Exporting our model to HDF5 file
nn.save('my_model.hdf5')
```

```python
# Trying to further create bins for the income amount column
income_value_counts = application_df['INCOME_AMT'].value_counts()
income_value_counts
```

```python
# Reading in the dataset again to include a new column
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")

# Making sure the bins are still present
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")

# Trying to further create bins for the application types column
application_types_to_replace = ['T9', 'T13', 'T12', 'T2', 'T25', 'T14', 'T29', 'T15', 'T17', 'T10']
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Making sure the bins are still present
for inc in income_value_to_replace:
    application_df['INCOME_AMT'] = application_df['INCOME_AMT'].replace(inc,"5M+")
    
```

```python
# Converting categorical data to numeric with `pd.get_dummies`
dummy_data = pd.get_dummies(data=application_df, columns=['APPLICATION_TYPE', 'AFFILIATION', 'CLASSIFICATION',
                                             'USE_CASE', 'ORGANIZATION', 'INCOME_AMT', 'IS_SUCCESSFUL'])
```

```python
# Splitting our preprocessed data into our features and target arrays
y = dummy_data["IS_SUCCESSFUL_1"].values
# Removing the Special considerations column to see if it enhances accuracy
X = dummy_data.drop(["IS_SUCCESSFUL_1", "IS_SUCCESSFUL_0", "SPECIAL_CONSIDERATIONS", "NAME"],1).values

# Splitting the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
# Creating a StandardScaler instances
scaler = StandardScaler()

# Fitting the StandardScaler
X_scaler = scaler.fit(X_train)

# Scaling the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Defining the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
hidden_nodes_layer1 = 80
hidden_nodes_layer2 = 120
hidden_nodes_layer3 = 100
hidden_nodes_layer4 = 50

nn2 = tf.keras.models.Sequential()

# First hidden layer
nn2.add(
    tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=40, activation="relu")
)

# Second hidden layer
nn2.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="relu"))

# Third hidden layer (changed activation function; added new layer)
nn2.add(tf.keras.layers.Dense(units=hidden_nodes_layer3, activation="tanh"))

# Fourth hidden layer (added new layer)
nn2.add(tf.keras.layers.Dense(units=hidden_nodes_layer4, activation="relu"))

# # Output layer
nn2.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Checking the structure of the model
nn2.summary()
```

```python
# Compiling the model
nn2.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Training the model (lesser epochs)
fit_model = nn2.fit(X_train_scaled,y_train,epochs=80)
```

```python
# Evaluating the model using the test data
model_loss, model_accuracy = nn2.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Exporting our model to HDF5 file
nn2.save('my_model2.hdf5')
```
